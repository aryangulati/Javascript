```python
import docx2txt as dt

```

```python
#load the data
#if you are on google colab you can do this method to choose the data on your PC
# from google.colab import files
# uploaded =files.upload()
```

```python
resume=dt.process("MARIA LISA Resume.docx")
print(resume)
```

```python
job=dt.process("Job_Resume.docx")
print(job)
```

```python
#list of text
text=[resume,job]
```

[ALso SEE THIS sklearn.feature_extraction.text]("sklearn.feature_extraction.text")

```python
#we have to create  a countvector object so we can get a count of each text 
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
count_matrix =cv.fit_transform(text)
```

[ALso SEE THIS from sklearn.metrics.pairwise import cosine_similarity ]("https://scikit-learn.org/stable/modules/generated/sklearn.metrics.pairwise.cosine_similarity.html#sklearn.metrics.pairwise.cosine_similarity")

```python
from sklearn.metrics.pairwise import cosine_similarity

#similarity score
print("\n Similarity Scores:")
print(cosine_similarity(count_matrix))

```

```python
Acceptedpercentage=cosine_similarity(count_matrix)[0][1]*100
Acceptedpercentage=round(Acceptedpercentage,2) #rounding it to 2 decimal places
print("Your Resume Accepted Percentage is "+str(Acceptedpercentage)+" % according to Job you applied for!!")
```
